```python
import pandas as pd
df=pd.read_csv('data.csv')
#data filtering
for x in df.index:
  if df.loc[x, 'Duration'] == 120:
    df.drop(x, inplace=True)
print(df)
for x in df.index:
  if df.loc[x, 'Duration'] > 120:
    df.drop(x, inplace=True)
print(df)
for x in df.index:
  if df.loc[x, 'Duration'] <60 :
    df.drop(x, inplace=True)

```

```python
#changed values
#if the value is higher than 120, set it to 120
import pandas as pd
df=pd.read_csv('data.csv')
for x in df.index:
  if df.loc[x, 'Duration'] > 120:
    df.loc[x, 'Duration']=120
for x in df.index:
  if df.loc[x, 'Duration']  > 60:
    df.loc[x, 'Duration']=400
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('data.csv')

# Assuming 'data' is your DataFrame and 'column_name' is the column containing the numerical data

# Step 1: Calculate Descriptive Statistics
descriptive_stats = df['Pulse'].describe()

# Step 2: Interquartile Range (IQR) Method
Q1 = descriptive_stats['25%']
Q3 = descriptive_stats['75%']
print(Q1)
print(Q3)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Identify Outliers
outliers = df[(df['Pulse'] < lower_bound) | (df['Pulse'] > upper_bound)]

# Step 4: Visualize the Data and Outliers
plt.figure(figsize=(10, 6))
plt.boxplot(df['Pulse'], vert=False)
plt.scatter(outliers['Pulse'], [1] * len(outliers), color='red', label='Outliers')
plt.legend()
plt.show()

# Print the outliers
print("Outliers:")
print(outliers)
print(len(outliers))
```
